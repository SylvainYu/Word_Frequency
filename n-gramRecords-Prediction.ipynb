```python
with open(stopwords_path, encoding='utf-8') as f:
    stopwords = f.readlines()
stopwords = set(map(lambda x:x.strip(), stopwords))  # 去除末尾换行符
len(stopwords)
```

```python
jieba.load_userdict('userdict.txt')  # 用户词典
word_table = {}  # 词表

        
# 全文统一处理
someDistrict = someDistrict.replace('丨', ' ')      # 替换特殊字符
someDistrict = someDistrict.replace(u'\t', '')      # 去除制表符
someDistrict = someDistrict.replace(u'\xa0', '')    # 去除全角空格
someDistrict = someDistrict.replace(u'\u3000', '')  # 去除不间断空白符
    
# 分割为句子的列表
someDistrict = re.split('，|。|；|？|！|：|\n', someDistrict)
someDistrict = list(filter(None, someDistrict))  # 去除空句子
    
print(len(someDistrict), 'sentences')

# 逐句处理并写入
file = open('./Records_data.txt', 'w')
for sentence in someDistrict:
    sentence = sentence.strip()
    cop = re.compile("[^\u4e00-\u9fa5]")
    sentence = cop.sub(' ', sentence)  # 将无效字符替换为空格
        
    # 分词、去除停止词
    sentence = jieba.cut(sentence)  # 分词
    word_list = [word.strip() for word in sentence if word.strip() and word not in stopwords]  # 去除停止词
    sentence = ' '.join(word_list)  # 用空格分隔分词结果
        
    for word in word_list:
        word = word.strip()
        word_table[word] = word_table.get(word, 0) + 1  # 频数加一
        
    if sentence:
        file.write(sentence + '\n')  # 写文件
file.close()
```

```python
# 保存词表到文件
with open('wordtable.txt', 'w', encoding='utf-8') as f:
    for i, word in enumerate(word_table):
        f.write(str(i) + ' ' + word + '\n')
```
